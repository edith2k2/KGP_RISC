```python
# Utilities
def to_bin5(n=0):
    i=2**4
    result=""
    while i>0:
        if n>=i:
            n-=i
            result+='1'
        else:
            result+='0'
        i//=2
    return result
def to_bin16(n=0):
    i=2**14
    result=""
    if n>=0:
        result+='0'
    else:
        n = 2**15+n
        result+='1'
    while i>0:
        if n>=i:
            n-=i
            result+='1'
        else:
            result+='0'
        i//=2
    return result
```
